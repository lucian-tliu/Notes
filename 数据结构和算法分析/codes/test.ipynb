```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(2,9)
n_fit = np.linspace(2,8,100)

def factorial(x):
    if x == 0 or x == 1:
        return 1
    else:
        return x * factorial(x - 1)

fig, ax = plt.subplots(figsize=(8,6),dpi=300)
ax.plot(n_fit, np.log(n_fit), label='log(n)', ls ='-')
# ax.plot(n_fit, np.sqrt(n_fit), label='sqrt(n)', ls ='--')
ax.plot(n_fit, n_fit, label='n', ls ='-.')
ax.plot(n_fit, n_fit*np.log(n_fit), label='nlog(n)', ls =':')
ax.plot(n_fit, n_fit**2, label='n^2', ls ='--')
ax.plot(n_fit, 2**n_fit, label='2^n', ls ='-.')
ax.plot(n, [factorial(i) for i in n], label='n!', marker='*') 
ax.set_yscale('log')
ax.set_xlim(2,8.5)
ax.set_xlabel('n', fontsize=14)
plt.legend()
plt.show()
```
