```python
from pylab import np, plt
from scipy.constants import hbar, e, m_e, epsilon_0, pi
from scipy.integrate import quad
```

```python
omega_fi = -5e9

P = lambda t, omega: np.abs((1-np.exp(1j*(omega_fi+omega)*t))/(omega_fi+omega)-(1-np.exp(1j*(omega_fi-omega)*t))/(omega_fi-omega))**2
```

```python
t=1e-6

omega = np.linspace(-6e9,6e9,1000)

P0 = P(t,omega)

Pi = P0 / np.max(P0)
```

```python
fig, ax = plt.subplots(figsize=(10,6),dpi=300)

ax.plot(omega*1e-9,Pi)
ax.vlines(np.abs(omega_fi)*1e-9,ymax=Pi.max()*1.9,ymin=0, color='red', linestyle='--', label=r'$\omega_{fi}$', linewidth=1)
ax.text(5.007, Pi.max()*0.9, r'$\omega_{fi}$', color='red', fontsize=14)
ax.set_xlabel(r'$\omega$ (GHz)', fontsize=14)
ax.set_ylim(0, Pi.max()*1.1)
ax.set_ylabel(r'$P_{i\to f}$ (a.u.)', fontsize=14)
ax.set_yticklabels([])
ax.set_yticks([])
ax.set_title(r'Transition Probability $P_{i\to f}(t)$ vs Driving Frequency $\omega$', fontsize=16)
plt.show()
```
